```python
# Create first network with Keras
from sklearn import preprocessing, cross_validation, neighbors
import pandas as pd
import numpy
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,confusion_matrix



def MLP(X,Y):
    X_train , X_test, y_train, y_test = cross_validation.train_test_split(X,Y,test_size=0.33)
    scaler = StandardScaler()
    scaler.fit(X_train)
    StandardScaler(copy=True, with_mean=True, with_std=True)
    
    # Now apply the transformations to the data:
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

    mlp = MLPClassifier(hidden_layer_sizes=(128,64,1),max_iter=500)
    mlp.fit(X_train,y_train)

    predictions = mlp.predict(X_test)
    print(classification_report(y_test,predictions))
```

```python
seed = 7
numpy.random.seed(seed)
dataset = numpy.loadtxt('/home/student/Downloads/data_cleaned.csv', delimiter=",")
X = dataset[:,0:110]
Y_ASOURCE = dataset[:,2]
Y_ATYPE = dataset[:,3]
Y_RACE = dataset[:,107]
Y_TOTCHG = dataset[:,109]
Y_ZIPINC_QRTL = dataset[:,110]
```

```python
MLP(X, Y_ASOURCE)
MLP(X, Y_ATYPE)
MLP(X, Y_RACE)
MLP(X, Y_TOTCHG)
MLP(X, Y_ZIPINC_QRTL)
```
